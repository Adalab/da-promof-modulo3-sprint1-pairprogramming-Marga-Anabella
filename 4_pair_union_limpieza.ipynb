# Ejercicios de limpieza

1. Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.


```python
df_merged.drop("country", axis=1, inplace=True)
```

```python
df_merged.head(1)
```

2. Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:
- No tengan espacios.
- Estén en minúscula.
- No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
- Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
- Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
new_columns = {columna: columna.lower().replace(" ", "").replace("(%)", "").replace("(km2)", "").replace(":", "").replace("\n", "").replace("(p/km2)", "") for columna in df_merged}
new_columns
```

```python
df_merged.rename(columns=new_columns, inplace=True)
df_merged.head(5)
```

3. La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.

```python
df_merged.info()
```

```python
df_merged["coordinates"].str.split(",", expand = True)
```

```python
df_merged[["lattitude", "longitude"]]= df_merged["coordinates"].str.split(",", expand = True)
df_merged.head(1)
```

```python
df_merged["lattitude"] = df_merged["lattitude"].str.replace(",", "").str.replace("(", "").str.replace("'", "")
```

```python
df_merged["longitude"] = df_merged["longitude"].str.replace(",", "").str.replace(")", "").str.replace("'", "")
```

```python
df_merged["longitude"]
```

```python
#eliminamos columna coordinates: 

df_merged.drop(columns="coordinates", axis=1, inplace=True)
```

```python
df_merged.head(1)
```

4. Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.

```python
# para una columna
df_merged["unemploymentrate"] = df_merged["unemploymentrate"].str.replace("%", "")
```

```python
#para iterar por las columnas que nos interesan 
columnas_con_perc = ["unemploymentrate", "taxrevenue", "totaltaxrate", "populationlaborforceparticipation", "outofpockethealthexpenditure", "grosstertiaryeducationenrollment", "grossprimaryeducationenrollment", "forestedarea", "cpichange", "agriculturalland"]
```

```python
for i in columnas_con_perc:
    df_merged[i] = df_merged[i].str.replace("%", "")
```

```python
df_merged.head(5)
```


5. Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".
